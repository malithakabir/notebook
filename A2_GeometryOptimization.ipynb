```python
import os
import py3Dmol
from rdkit import Chem
from rdkit.Chem import AllChem
from ipywidgets import interact, interactive, fixed
```

```python
m = Chem.MolFromSmiles(r'COc1ccc2[nH]c([S@@+]([O-])Cc3ncc(C)c(OC)c3C)nc2c1') # esomeprazole
m = Chem.AddHs(m)
AllChem.EmbedMultipleConfs(m,numConfs=10)
# align to one of the ring systems:
AllChem.AlignMolConformers(m,m.GetSubstructMatch(Chem.MolFromSmarts('c1[nH]c2ccccc2n1')))
```

Before energy minimization

```python
print(Chem.MolToMolBlock(m,confId=1))
```

```python
ff = AllChem.UFFGetMoleculeForceField(m,confId=1)

maxCycles=10

while maxCycles:
    if ff.Minimize():
        maxCycles -= 1
    else:
        break

```

```python
#dir(ff)
```

```python
ff.CalcEnergy()
```

```python
ff.Minimize()
```

```python
print(Chem.MolToMolBlock(m,confId=1))
```
